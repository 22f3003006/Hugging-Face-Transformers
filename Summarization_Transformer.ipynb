```python
import gradio as gr
from transformers import pipeline
from PyPDF2 import PdfReader
```

```python
r = PdfReader('/content/PRANAV JOSHI - 22F3003006 - IITM BS.pdf')
print(r.pages[0].extract_text())
```

```python
model = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarizer(input,length_choice,file):
  if length_choice == "small":
        min_len, max_len = 15, 50
  elif length_choice == "large":
        min_len, max_len = 80, 200
  else:
        min_len, max_len = 30, 130
  if file:
    r = PdfReader(file)
    text = ""
    for page in r.pages:
      text += page.extract_text()
    input = text
  print(input)
  return model(input,min_length=min_len,max_length=max_len,do_sample=False)[0]['summary_text']

iface = gr.Interface(
    fn=summarizer,
    inputs=[
        gr.Textbox(placeholder="Something you wanna summarize!"),
        gr.Radio(["small","medium","large"],label="Summary Length",value="medium"),
        gr.File(label="Upload PDF", file_types=[".pdf"])
    ],
    outputs=gr.Textbox(label="Summary",scale=7, lines=8),
    title="Text Summarizer <3",
    description="Enter a text to be summarized using FaceBook BART!")
iface.launch(debug=True)
```
